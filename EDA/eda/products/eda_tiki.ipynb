### 2. Data Loading

```python
import os
import pandas as pd

file_names = ["dien_thoai.csv", "may_tinh_linh_kien.csv","phu_kien.csv"]

dataframes = []
categories = ["Điện thoại","Laptop - Máy tính và Linh kiện", "Phụ kiện số"]

dir = 'E:\MY CAREER\Internship Project\EDA\dataset\\tiki'

for file_name, category in zip(file_names, categories):
    path = os.path.join(dir, file_name)
    try:
        df = pd.read_csv(path)
        df["category"] = category
        dataframes.append(df)
    except pd.errors.ParserError as e:
        print(f"Error reading file {file_name}: {e}")

combined_df = pd.concat(dataframes, ignore_index=True)
combined_df.head()
```

```python
combined_df.columns
```

```python
new_df = combined_df[['id', 'name','sku', 'url_key','brand_name', 'seller_product_id','seller_id','price','original_price', 'discount','discount_rate',
       'rating_average', 'review_count', 'productset_id', 'category', 'quantity_sold']]
new_df.info()
```

### 3. Data Cleaning & Exploring

```python
dup = new_df.duplicated(subset='id').sum()
dup
```

```python
new_df.drop_duplicates(subset='id',inplace=True, keep='last')
```

```python
m = new_df['brand_name'].unique().tolist()
m.sort()
print(m)
```

```python
new_df['brand_name'] = new_df['brand_name'].str.strip()
len(new_df['brand_name'].unique())
```

```python
new_df['brand_name'] = new_df['brand_name'].str.upper()
len(new_df['brand_name'].unique())
```

```python
brand_mapping = {
    '5 TECH': '5TECH',
    'BKAV PRO': 'BKAV',
    'COOLERMASTER': 'COOLER MASTER',
    'CYBERPOWER': 'CYBER POWER',
    'DORON THE BEST SERVICE': 'DORON',
    'GIGABYE': 'GIGABYTE',
    'HUMANMOTION': 'HUMAN MOTION',
    'ITEL': 'INTEL',
    'KAI N': 'KAIN',
    'LENOVO FLAGSHIP': 'LENOVO',
    'LENOVO IDEAPAD': 'LENOVO',
    'LENOVO THINKPAD': 'LENOVO',
    'LEXAR (Mỹ)': 'LEXAR',
    'MASSEL': 'MASSTEL',
    'MAXCELL': 'MAXELL',
    'TEAM GROUP': 'TEAM',
    'TP-LINK': 'TPLINK'
}
new_df['brand_name'] = new_df['brand_name'].replace(brand_mapping)
```

```python
len(new_df['brand_name'].unique())
```

```python
def extract_value(s):
    try:
        return eval(s)['value']
    except (SyntaxError, TypeError):
        return None

# Apply the extraction function to 'quantity_sold' column
new_df['sold'] = new_df['quantity_sold'].apply(extract_value)
```

```python
new_df['sold'].fillna(0, inplace=True)
new_df['sold'] = new_df['sold'].astype(int)
new_df
```

```python
new_df.drop('quantity_sold', axis=1, inplace=True)

new_df
```

```python
new_df.to_csv('cleandata/tiki_product.csv')
```
