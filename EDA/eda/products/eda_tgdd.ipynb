### 2. Data Loading

```python
import os
import pandas as pd

file_names = ["dien_thoai.csv", "may_tinh_bang.csv","laptop.csv", "dong_ho_thong_minh.csv"]

dataframes = []
categories = ["Điện thoại","Máy tính bảng", "Laptop", "Đồng hồ thông minh"]


dir = 'E:\MY CAREER\Internship Project\EDA\dataset\\tgdd\products'
for file_name, category in zip(file_names, categories):
    path = os.path.join(dir, file_name)
    try:
        df = pd.read_csv(path)
        df["category"] = category  # Thêm cột "category" với giá trị tương ứng
        dataframes.append(df)
    except pd.errors.ParserError as e:
        print(f"Error reading file {file_name}: {e}")

combined_df = pd.concat(dataframes, ignore_index=True)
combined_df.head()
```

```python
combined_df.columns
```

```python
new_df = combined_df[['id', 'title', 'link', 'description', 'box_price_present',
       'box_price_old', 'box_price_percent', 'rating', 'cmt_count',
       'category']]
new_df.info()
```

##### Handling missing value

Đối với giá có 2 có thể biểu hiện giá là 'box_price_present' biểu hiện giá sản phẩm hiện tại và 'box_price_old' biểu hiện giá sản phẩm trước đây (khi chưa áp dụng discount,...)
Có 2 TH có thể xảy ra:
+ Cả 2 cột này đều không có giá trị: Chỉ có thể xóa
+ Cột box_price_old thiếu thì dựa trên trường 'box_price_present'

```python
new_df.dropna(subset=['box_price_present', 'box_price_old'], how='all', inplace=True)
```

```python
new_df.info()
```

+ Chuẩn hóa giá về kiểu số
+ Điền các giá trị còn thiếu của 'box_price_old' là 0 

```python
def normalize_price(price):
    price = ''.join(c for c in price if c.isdigit() or c == ',')
    price = price.replace(',', '').replace('₫', '')
    try:
        price = int(price)
        return price
    except ValueError:
        return None
    

new_df['box_price_present'] = new_df['box_price_present'].apply(normalize_price)
```

```python
new_df['box_price_old'] = new_df['box_price_old'].str.replace('.', '')
new_df['box_price_old'] = new_df['box_price_old'].str.replace('₫', '')
new_df['box_price_old'].fillna(0,inplace=True)
new_df['box_price_old'] = new_df['box_price_old'].astype('int')

new_df 
```

```python
new_df['box_price_present'].isna().sum()
```

```python
new_df['box_price_percent'] = new_df['box_price_percent'].str.replace('%','')
new_df['box_price_percent'] = pd.to_numeric(new_df['box_price_percent'], errors='coerce')

new_df['box_price_percent'].fillna(0, inplace=True)
```

```python
zero_price_sum = new_df[new_df['box_price_old'] == 0]['box_price_old']
len(zero_price_sum)
```

```python
zero_price_sum = new_df[new_df['box_price_percent'] == 0]['box_price_percent']
len(zero_price_sum)
```

```python
mask = new_df['box_price_old'] == 0
new_df.loc[mask, 'box_price_old'] = new_df.loc[mask, 'box_price_present']
```

```python
new_df['discount'] = ((new_df['box_price_old'] - new_df['box_price_present']) * 100/ new_df['box_price_old'])
```

```python
new_df['discount'] = new_df['discount'].astype('int')
new_df
```
